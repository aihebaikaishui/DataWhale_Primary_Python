## Pandas plot 出图
——————这节使用的matpotlib仅仅是用来show图片的
* ### 创建一个series
* ### dataframe 可视化

* ### 创建一个series

```python
import numpy as np
import pandas as  pd
import matplotlib.pyplot as plt
#随机生成1000个数据
data = pd.Series(np.random.randn(1000),index=np.arange(1000))
# 为了方便观看效果，累加数据
data.cumsum()
#pandas数据直接观看
data.plot()
plt.show()
```

* ### dataframe 可视化

```python
# 生成一个1000*4的DateFrame,
data = pd.DataFrame(
    np.random.randn(1000,4),
    index=np.arange(1000),
    columns=list("ABCD")
    )
data.cumsum()
data.plot() # 默认的为连续的
plt.show()
```
